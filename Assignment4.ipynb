```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("/Airbnb Dataset 19.csv")
```

```python
data.head()
```

**Histogram of prices**

```python
plt.figure(figsize=(10, 6))
plt.hist(data['price'], bins=30, edgecolor='black')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Histogram of Airbnb Listing Prices')
plt.show()
```

Bar Plot of Room Types:


```python
plt.figure(figsize=(10, 6))
sns.countplot(x='room_type', data=data)
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.title('Bar Plot of Room Types')
plt.xticks(rotation=45)
plt.show()



```

Box Plot of Prices by Neighborhood:

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='neighbourhood_group', y='price', data=data)
plt.xlabel('neighbourhood')
plt.ylabel('Price')
plt.title('Box Plot of Prices by neighborhood')
plt.xticks(rotation=45)
plt.show()
```

**Heatmap of Correlation:**
Purpose of a Correlation Heatmap:
A correlation heatmap is useful for identifying relationships between variables. It can help you understand which variables tend to move together (positive correlation) or move in opposite directions (negative correlation). This insight is crucial for feature selection, data preprocessing, and understanding the underlying structure of your data.

```python
correlation_matrix = data.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap of Correlation')
plt.show()
```

**Scatter Plot of Prices vs. Ratings:**

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='price', y='reviews_per_month', data=data)
plt.xlabel('Price')
plt.ylabel('reviews_per_month')
plt.title('Scatter Plot of Prices vs.reviews_per_month')
plt.show()
```

**Pie Chart of Room Type Distribution:**

```python
room_type_counts = data['room_type'].value_counts()
plt.pie(room_type_counts, labels=room_type_counts.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Room Type Distribution')
plt.show()
```
